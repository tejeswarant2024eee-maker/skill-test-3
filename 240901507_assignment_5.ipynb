```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder

# Data provided by the user
data = {
    'Transmitter': ['T1', 'T2', 'T1', 'T3', 'T2', 'T1', 'T3', 'T2', 'T1', 'T3'],
    'Signal Strength': [75, 50, 90, 40, 85, 70, 30, 60, 80, 20],
    'Frequency': [2000, 1500, 2200, 1300, 2100, 1800, 1200, 1600, 1900, 1100],
    'Signal Quality': ['Good', 'Bad', 'Good', 'Bad', 'Good', 'Good', 'Bad', 'Bad', 'Good', 'Bad']
}
df_signal = pd.DataFrame(data)

# Encode categorical 'Transmitter' feature using one-hot encoding
df_signal = pd.get_dummies(df_signal, columns=['Transmitter'], drop_first=True)

# Encode target variable 'Signal Quality'
le = LabelEncoder()
df_signal['Signal Quality_Encoded'] = le.fit_transform(df_signal['Signal Quality'])

display(df_signal)
```

The data has been loaded, categorical features (Transmitter) have been one-hot encoded, and the target variable (Signal Quality) has been label encoded ('Bad': 0, 'Good': 1). Now, I will define the features (X) and the target (y), split the data into training and testing sets, and then train a Decision Tree Classifier.

```python
# Define features (X) and target (y)
X = df_signal.drop(['Signal Quality', 'Signal Quality_Encoded'], axis=1)
y = df_signal['Signal Quality_Encoded']

# Split data into training and testing sets
# Using a test_size of 0.3 to ensure reasonable splits for both classes
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Initialize and train the Decision Tree Classifier
dt_classifier = DecisionTreeClassifier(random_state=42)
dt_classifier.fit(X_train, y_train)

print("Decision Tree Classifier trained successfully!")
```

The Decision Tree Classifier has been trained. Next, I will make predictions on the test set and evaluate the model's performance using accuracy, a classification report, and a confusion matrix.

```python
# Make predictions on the test set
y_pred = dt_classifier.predict(X_test)

# Evaluate the model
print(f"Accuracy: {accuracy_score(y_test, y_pred):.2f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```
