```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Updated data provided by the user
data_updated = {
    'Voltage (V)': [5, 5.5, 4.8, 6, 6.5, 6.2, 4.9, 5.8, 5.2, 6.3],
    'Current (A)': [0.5, 0.4, 0.6, 0.3, 0.2, 0.25, 0.55, 0.35, 0.5, 0.22],
    'Status': ['Working', 'Working', 'Working', 'Defective', 'Defective', 'Defective', 'Working', 'Defective', 'Working', 'Defective']
}
df_updated = pd.DataFrame(data_updated)

# Encode labels: Working -> 1, Defective -> 0
# The original request specified 'Defective -> 1', but assigning both 'Working' and 'Defective' to 1
# would make classification impossible as both classes would be indistinguishable.
# Therefore, I will use Working -> 1 and Defective -> 0 for a meaningful binary classification.
df_updated['Status_Encoded'] = df_updated['Status'].map({'Working': 1, 'Defective': 0})

display(df_updated)
```

The updated data has been loaded, and the 'Status' column has been encoded with 'Working' as `1` and 'Defective' as `0`. Next, I will define the features (X) and the target (y), split the data into training and testing sets, and then train a KNN classifier.

```python
# Define features (X) and target (y)
X_updated = df_updated[['Voltage (V)', 'Current (A)']]
y_updated = df_updated['Status_Encoded']

# Split data into training and testing sets
# With 5 instances of each class, a test_size=0.3 (3 samples) should work for stratification
# This results in 2 samples of each class in the test set (if rounding allows) or 1 'Defective' and 2 'Working'.
# Let's try test_size=0.2 to ensure at least 1 sample from each class in test set and sufficient for training.
X_train, X_test, y_train, y_test = train_test_split(X_updated, y_updated, test_size=0.2, random_state=42, stratify=y_updated)

# Initialize and train the KNN classifier
# Choosing k=3 as it's a small dataset, but k should be tuned for larger datasets.
# A k value of 1 might overfit. With 10 samples and 2 in the test set, k=3 seems reasonable.
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

print("KNN model trained successfully!")
```

The KNN model has been trained. Now, I will make predictions on the test set and evaluate the model's performance using accuracy and a classification report.

```python
# Make predictions on the test set
y_pred = knn.predict(X_test)

# Evaluate the model
print(f"Accuracy: {accuracy_score(y_test, y_pred):.2f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```
